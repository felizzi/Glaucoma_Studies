```python
## extract the latest portuguese mortality
import pandas as pd
df_male = pd.read_csv("data/mltper_1x1.txt", delim_whitespace=True, skiprows=1)
df_female = pd.read_csv("data/fltper_1x1.txt", delim_whitespace=True, skiprows=1)

## select only the latest year available, Year == max(Year)

df_male_latest = df_male[df_male['Year'] == df_male['Year'].max()]
df_female_latest = df_female[df_female['Year'] == df_female['Year'].max()]  

## transform qx column to numeric
df_male_latest['qx'] = pd.to_numeric(df_male_latest['qx'], errors='coerce')
df_female_latest['qx'] = pd.to_numeric(df_female_latest['qx'], errors='coerce')

# convert Age = 110+ to 110
df_male_latest.loc[df_male_latest['Age'] == '110+', 'Age'] = 110
df_female_latest.loc[df_female_latest['Age'] == '110+', 'Age'] = 110

df_male_latest['Age'] = pd.to_numeric(df_male_latest['Age'], errors='coerce')
df_female_latest['Age'] = pd.to_numeric(df_female_latest['Age'], errors='coerce')

## save the latest datasets
df_male_latest.to_csv("data/male_mortality_2023.csv", index=False)
df_female_latest.to_csv("data/female_mortality_2023.csv", index=False)  
```
